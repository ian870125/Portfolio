```python
df = df[~(df['InvoiceNo'].str.contains('A'))] #去除呆帳
df = df[df.Quantity > 0]
```

```python
import nltk
df['cut_word'] = df['Description'].astype(str).apply(lambda x: nltk.tokenize.RegexpTokenizer(r'\w+').tokenize(x)) #斷詞

```

```python
wnl = nltk.stem.WordNetLemmatizer() # 詞性還原
def lem(text):
    for i in range(len(text)):
        text[i] = wnl.lemmatize(text[i])
    return text
df['cut_word'] = df['cut_word'].apply(lem)
```

```python
import string #去除標點符號、停用詞
def remove_stopword(text):
    stopword_list = set(nltk.corpus.stopwords.words('english') 
                    + list(string.punctuation))
    tokens = [token.strip().lower() for token in text]
    filtered_tokens = [token for token in tokens if token not in stopword_list]
    return filtered_tokens
df['cut_word'] = df['cut_word'].apply(remove_stopword)
```

```python
import collections
def combine(text):
    word_freqs = collections.Counter()
    for i in text:
        for j in i:
            word_freqs[j] += 1
    
    return word_freqs.most_common(5)
bow_country = df.groupby('Country')['cut_word'].apply(combine) #各國常見詞
```

```python
bow_country['United Kingdom']
bow_country['Saudi Arabia']
```

```python
def concat(text):
    conc = []
    for i in text:
        conc.append(i)
    return conc
Corpus = nltk.text.TextCollection(df['cut_word'])
```

```python
def TF_IDF(Corpus, WordList):
    TF_IDF_Dict = {}
    for Word in WordList:
        TF_IDF_Dict[Word] = Corpus.tf_idf(Word, WordList)  
    return TF_IDF_Dict
```

```python
import warnings
warnings.filterwarnings('ignore')
tf_idf = pd.DataFrame(index = df.index)
for i in df.index:
    t_i = TF_IDF(Corpus, df.loc[i, 'cut_word'])
    for j in t_i:
        tf_idf.loc[i, j] = t_i[j]
tf_idf
```

```python
tf_idf['y'] = df['Quantity'] * df['UnitPrice']
tf_idf.head()
```

```python
tf_idf_na = tf_idf.isna().sum()
st = np.std(tf_idf_na[:-1])
reserve_var = tf_idf_na[tf_idf_na < (np.mean(tf_idf_na) - 3 * st)].index#去除na過大
```

```python
tf_idf = tf_idf.loc[:, reserve_var]
tf_idf.fillna(0, inplace = True)
from sklearn import tree
from sklearn.model_selection import train_test_split
X, Y = tf_idf.iloc[:, :-1], tf_idf.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
tree_model = tree.DecisionTreeRegressor(max_depth=5, 
                                        min_samples_leaf=1,
                                        ccp_alpha=0.0, 
                                        random_state=111)
y_predict = tree_model.fit(X_train, y_train)
```

```python
y_predict = tree_model.predict(X_train)
np.sum((y_predict - y_train) ** 2) 
```

```python
plt.figure(figsize=(30, 30))
tree.plot_tree(y_predict, 
                   feature_names=X.columns,  
                   class_names='target',
                   filled=True)
```

```python
import datetime as dt #RFM R
df.dropna(subset=['CustomerID'],how='all',inplace=True)
now = dt.date(2011,12,9)
df['date'] = pd.DatetimeIndex(df.InvoiceDate).date
recency_df = df.groupby(['CustomerID'],as_index=False)['date'].max()
recency_df.columns = ['CustomerID','LastPurchaseDate']
recency_df['Recency'] = recency_df.LastPurchaseDate.apply(lambda x : (now - x).days)
recency_df.drop(columns=['LastPurchaseDate'],inplace=True)
```

```python
frequency_df = df.copy() # RFM F
frequency_df.drop_duplicates(subset=['CustomerID','InvoiceNo'], keep="first", inplace=True) 
frequency_df = frequency_df.groupby('CustomerID',as_index=False)['InvoiceNo'].count()
frequency_df.columns = ['CustomerID','Frequency']
frequency_df.head()

```

```python
df['Total_cost'] = df['UnitPrice'] * df['Quantity'] # RFM M
monetary_df=df.groupby('CustomerID',as_index=False)['Total_cost'].sum()
monetary_df.columns = ['CustomerID','Amount']
```

```python
rf = recency_df.merge(frequency_df,left_on='CustomerID',right_on='CustomerID') #RFM
rfm = rf.merge(monetary_df,left_on='CustomerID',right_on='CustomerID')
rfm.set_index('CustomerID',inplace=True)
rfm.head()
```

```python
quantiles = rfm.quantile(q=[0.33,0.66])
def R_Class(x,p,d):
    if x <= d[p][0.33]:
        return 3
    elif x <= d[p][0.66]:
        return 2
    else:
        return 1
    
def FM_Class(x,p,d):
    if x <= d[p][0.33]:
        return 1
    elif x <= d[p][0.66]:
        return 2
    else:
        return 3
```

```python
RFM_Segment = rfm.copy()
RFM_Segment['R_Quartile'] = RFM_Segment['Recency'].apply(R_Class, args=('Recency',quantiles,))
RFM_Segment['F_Quartile'] = RFM_Segment['Frequency'].apply(FM_Class, args=('Frequency',quantiles,))
RFM_Segment['M_Quartile'] = RFM_Segment['Amount'].apply(FM_Class, args=('Amount',quantiles,))

RFM_Segment['RFMClass'] = RFM_Segment.R_Quartile.map(str) \
                            + RFM_Segment.F_Quartile.map(str) \
                            + RFM_Segment.M_Quartile.map(str)

for i in range(1, 4):
    for j in range(1, 4):
        for  k in range(1, 4):
            print('{}: {}'.format([i, j, k], len(RFM_Segment[RFM_Segment['RFMClass']== str(i) + str(j) + str(k)])/len(RFM_Segment)))
```
